```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.utils import normalize

def load_data(image_dir, mask_dir, target_size=(256, 256)):
    images = []
    masks = []
    image_files = [f for f in sorted(os.listdir(image_dir)) if f.endswith('.png')]
    mask_files = [f for f in sorted(os.listdir(mask_dir)) if f.endswith('.png')]
    
    for image_file, mask_file in zip(image_files, mask_files):
        image_path = os.path.join(image_dir, image_file)
        mask_path = os.path.join(mask_dir, mask_file)

        image = load_img(image_path, target_size=target_size)
        mask = load_img(mask_path, color_mode="grayscale", target_size=target_size)

        image = img_to_array(image)
        mask = img_to_array(mask)

        mask = np.where(mask == 255, 1, 0)  # Convert to binary mask

        images.append(image)
        masks.append(mask)

    images = np.array(images, dtype="float32") / 255.0  # Normalize images
    masks = np.array(masks, dtype="float32")

    return images, masks

image_dir = 'data_road/processed_data/images'
mask_dir = 'data_road/processed_data/greyscale_scaled_255_masks'
images, masks = load_data(image_dir, mask_dir)
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Conv2DTranspose, concatenate, Input
from tensorflow.keras.models import Model

def unet_model(input_size=(256, 256, 3)):
    inputs = Input(input_size)
    
    # Downsample
    c1 = Conv2D(16, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inputs)
    p1 = MaxPooling2D((2, 2))(c1)
    # Add more layers as needed here for a deeper network

    # Upsample and concatenate
    u = Conv2DTranspose(16, (3, 3), strides=(2, 2), padding='same')(p1)
    u = concatenate([u, c1])
    c = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(u)

    model = Model(inputs=[inputs], outputs=[c])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = unet_model()
model.summary()
```

```python
model.fit(images, masks, batch_size=2, epochs=100, validation_split=0.1)
```

```python
model.save('lane_detection_model.h5')
```

```python
import cv2
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
import numpy as np

# Load the saved model
model = load_model('lane_detection_model.h5')

def predict_on_image(image_path, model_input_size=(256, 256)):
    original_image = load_img(image_path)
    original_size = original_image.size  # PIL Image.size gives (width, height)
    original_array = img_to_array(original_image)

    # Resize for model input
    img = cv2.resize(original_array, model_input_size)
    img = img / 255.0
    img = np.expand_dims(img, axis=0)
    
    # Model prediction
    prediction = model.predict(img)[0, :, :, 0]
    
    # Resize prediction to original image size
    prediction_resized = cv2.resize(prediction, (original_size[0], original_size[1]), interpolation=cv2.INTER_LINEAR)
    
    # Create a superimposed image
    mask = (prediction_resized * 255).astype(np.uint8)  # Convert mask to uint8 format
    heatmap = cv2.applyColorMap(mask, cv2.COLORMAP_JET)  # Apply a color map to mask
    superimposed_img = cv2.addWeighted(original_array.astype(np.uint8), 0.6, heatmap, 0.4, 0)
    
    return original_image, prediction_resized, superimposed_img, mask

# Path to the image you want to predict on
image_path = 'data_road/processed_data/images/000076.png'
original_image, predicted_mask, superimposed_image, mask = predict_on_image(image_path)

# Save the predicted mask and superimposed image
cv2.imwrite('predicted_mask.png', mask)
cv2.imwrite('superimposed_image.png', superimposed_image)

# Display the results
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(predicted_mask, cmap='gray')
plt.title('Predicted Mask')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(superimposed_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title('Superimposed Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array

# Load the saved model
model = load_model('lane_detection_model.h5')

def process_frame(frame):
    # Resize the frame to match the model's expected input
    input_size = (256, 256)
    frame_resized = cv2.resize(frame, input_size)
    frame_normalized = frame_resized / 255.0
    frame_normalized = np.expand_dims(frame_normalized, axis=0)
    
    # Predict the mask
    prediction = model.predict(frame_normalized)[0, :, :, 0]
    prediction_resized = cv2.resize(prediction, (frame.shape[1], frame.shape[0]), interpolation=cv2.INTER_LINEAR)
    
    # Create a superimposed image
    mask = (prediction_resized * 255).astype(np.uint8)
    heatmap = cv2.applyColorMap(mask, cv2.COLORMAP_JET)
    superimposed_img = cv2.addWeighted(frame, 0.6, heatmap, 0.4, 0)
    
    return superimposed_img

def process_video(input_video_path, output_video_path):
    cap = cv2.VideoCapture(input_video_path)
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = None
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Process the frame
        processed_frame = process_frame(frame)
        
        # Initialize video writer
        if out is None:
            out = cv2.VideoWriter(output_video_path, fourcc, 20.0, (frame.shape[1], frame.shape[0]))
        
        # Write the frame to the output video
        out.write(processed_frame)
    
    cap.release()
    if out:
        out.release()
    print("Video processing complete. Output saved as:", output_video_path)

# Specify the paths to your input video and where you want to save the output
input_video_path = 'video_test/challenge.mp4'
output_video_path = 'video_test/challenge_inference.avi'

# Process the video
process_video(input_video_path, output_video_path)
```
