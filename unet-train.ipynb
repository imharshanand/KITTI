```python
# Import necessary libraries
import os
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam

# Define paths for training images and masks
train_images_path = "data_road/training/image_2/"
train_masks_path = "data_road/training/gt_image_2/"
test_images_path = "data_road/testing/image_2/"
```

```python
# Function to load images and masks based on matching filenames
def load_images_and_masks(image_path, mask_path, num_samples=10):
    image_files = os.listdir(image_path)
    mask_files = os.listdir(mask_path)
    
    image_map = {file.split('_')[-1]: file for file in image_files}
    mask_map = {file.split('_')[-1]: file for file in mask_files}
    
    common_keys = set(image_map.keys()).intersection(set(mask_map.keys()))
    common_keys = list(common_keys)
    
    if len(common_keys) == 0:
        print("No common files found between images and masks.")
        return np.array([]), np.array([]), [], []
    
    random.shuffle(common_keys)
    num_samples = min(num_samples, len(common_keys))
    selected_keys = common_keys[:num_samples]
    
    images, masks, image_filenames, mask_filenames = [], [], [], []
    for key in selected_keys:
        img_file = image_map[key]
        mask_file = mask_map[key]
        
        img = cv2.imread(os.path.join(image_path, img_file))
        img = cv2.resize(img, (256, 256))
        images.append(img)
        image_filenames.append(img_file)
        
        mask = cv2.imread(os.path.join(mask_path, mask_file), cv2.IMREAD_GRAYSCALE)
        mask = cv2.resize(mask, (256, 256))
        masks.append(mask)
        mask_filenames.append(mask_file)
    
    return np.array(images), np.array(masks), image_filenames, mask_filenames

# Load training images and masks
train_images, train_masks, image_filenames, mask_filenames = load_images_and_masks(train_images_path, train_masks_path, num_samples=50)

# Check if images and masks were loaded correctly
if train_images.size == 0 or train_masks.size == 0:
    print("Error: No images or masks were loaded.")
else:
    print(f"Loaded {train_images.shape[0]} images and masks.")
```

```python
# Normalize images
train_images = train_images / 255.0

# Function to visualize a few samples with filenames
def visualize_samples_with_filenames(images, masks, image_filenames, mask_filenames, num_samples=5):
    for i in range(num_samples):
        plt.figure(figsize=(10, 5))

        plt.subplot(1, 2, 1)
        plt.imshow(images[i])
        plt.title(f'Image: {image_filenames[i]}')

        plt.subplot(1, 2, 2)
        plt.imshow(masks[i], cmap='gray')
        plt.title(f'Mask: {mask_filenames[i]}')

        plt.show()

# Visualize samples
visualize_samples_with_filenames(train_images, train_masks, image_filenames, mask_filenames, num_samples=5)
```

```python
# Function to extract unique values and their counts from the masks
def extract_unique_values_and_counts(masks):
    unique_values, counts = np.unique(masks, return_counts=True)
    unique_counts = list(zip(unique_values, counts))
    unique_counts.sort(key=lambda x: x[1], reverse=True)
    return unique_counts

unique_counts = extract_unique_values_and_counts(train_masks)

# Print unique values and their counts
print("Unique values and their counts (sorted in descending order):")
for value, count in unique_counts:
    print(f'Value: {value}, Count: {count}')

# Assign names to the top 3 features
top_features = unique_counts[:3]
feature_names = {
    76: "background",
    105: "drive",
    0: "opposite"
}

print("Available features:")
for value, name in feature_names.items():
    print(f"Value: {value}, Name: {name}")

# Select the features for the model to detect
selected_features = ["drive"]

# Prepare the masks for training
num_classes = len(selected_features)
train_labels = np.zeros((train_masks.shape[0], 256, 256, num_classes))

for i, (value, name) in enumerate(feature_names.items()):
    if name in selected_features:
        train_labels[..., selected_features.index(name)] = (train_masks == value).astype(np.uint8)

print(f"Train labels shape: {train_labels.shape}")

print(f"Total number of images: {train_images.shape[0]}")
print(f"Total number of masks: {train_masks.shape[0]}")
print(f"Image dimensions: {train_images.shape[1:4]}")
print(f"Mask dimensions: {train_masks.shape[1:3]}")
```

```python
# Function to visualize more final images and masks
def visualize_final_images_and_masks(images, masks, num_samples=5):
    for i in range(num_samples):
        plt.figure(figsize=(10, 5))

        plt.subplot(1, 2, 1)
        plt.imshow(images[i])
        plt.title(f'Final Image {i+1}')

        plt.subplot(1, 2, 2)
        plt.imshow(masks[i, ..., 0], cmap='gray')
        plt.title(f'Final Mask {i+1}')

        plt.show()

# Visualize final images and masks
visualize_final_images_and_masks(train_images, train_labels, num_samples=5)
```

```python
# Define the UNet model for binary segmentation
def unet_model(input_size=(256, 256, 3)):
    inputs = Input(input_size)
    # Downsampling
    c1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    c1 = Conv2D(64, 3, activation='relu', padding='same')(c1)
    p1 = MaxPooling2D(pool_size=(2, 2))(c1)

    c2 = Conv2D(128, 3, activation='relu', padding='same')(p1)
    c2 = Conv2D(128, 3, activation='relu', padding='same')(c2)
    p2 = MaxPooling2D(pool_size=(2, 2))(c2)

    c3 = Conv2D(256, 3, activation='relu', padding='same')(p2)
    c3 = Conv2D(256, 3, activation='relu', padding='same')(c3)
    p3 = MaxPooling2D(pool_size=(2, 2))(c3)

    c4 = Conv2D(512, 3, activation='relu', padding='same')(p3)
    c4 = Conv2D(512, 3, activation='relu', padding='same')(c4)
    p4 = MaxPooling2D(pool_size=(2, 2))(c4)

    # Bottleneck
    c5 = Conv2D(1024, 3, activation='relu', padding='same')(p4)
    c5 = Conv2D(1024, 3, activation='relu', padding='same')(c5)

    # Upsampling
    u6 = UpSampling2D(size=(2, 2))(c5)
    u6 = concatenate([u6, c4], axis=-1)
    c6 = Conv2D(512, 3, activation='relu', padding='same')(u6)
    c6 = Conv2D(512, 3, activation='relu', padding='same')(c6)

    u7 = UpSampling2D(size=(2, 2))(c6)
    u7 = concatenate([u7, c3], axis=-1)
    c7 = Conv2D(256, 3, activation='relu', padding='same')(u7)
    c7 = Conv2D(256, 3, activation='relu', padding='same')(c7)

    u8 = UpSampling2D(size=(2, 2))(c7)
    u8 = concatenate([u8, c2], axis=-1)
    c8 = Conv2D(128, 3, activation='relu', padding='same')(u8)
    c8 = Conv2D(128, 3, activation='relu', padding='same')(c8)

    u9 = UpSampling2D(size=(2, 2))(c8)
    u9 = concatenate([u9, c1], axis=-1)
    c9 = Conv2D(64, 3, activation='relu', padding='same')(u9)
    c9 = Conv2D(64, 3, activation='relu', padding='same')(c9)

    outputs = Conv2D(1, 1, activation='sigmoid')(c9)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
    
    return model

model = unet_model()

# Define a model checkpoint callback to save the best model
checkpoint = ModelCheckpoint('unet_model.h5', monitor='val_loss', verbose=1, save_best_only=True, mode='min')

# Train the model
history = model.fit(train_images, train_labels, validation_split=0.1, epochs=100, batch_size=4, callbacks=[checkpoint])

# Save the final model
model.save('unet_model_final.h5')
```

```python
# Function to load and preprocess a test image for inference
def load_and_preprocess_test_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (256, 256))
    img = img / 255.0
    img = np.expand_dims(img, axis=0)  # Add batch dimension
    return img

# Load a test image
test_image_path = os.path.join(test_images_path, os.listdir(test_images_path)[0])
test_image = load_and_preprocess_test_image(test_image_path)

# Perform inference
prediction = model.predict(test_image)
predicted_mask = (prediction[0, ..., 0] > 0.5).astype(np.uint8)  # Thresholding the sigmoid output

# Visualize the test image and its predicted mask
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(test_image[0])
plt.title('Test Image')

plt.subplot(1, 2, 2)
plt.imshow(predicted_mask, cmap='gray')
plt.title('Predicted Mask')

plt.show()
```
